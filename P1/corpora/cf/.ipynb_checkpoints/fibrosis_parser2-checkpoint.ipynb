```python
import re

f = open("cf79","r")
data = f.read()
f.close()

js = open("dat.json","w") 
js.write("{\n\n")
exp="(PN (.|\n|\r\n)*?)\n\n"

for d in re.findall(exp,data):
    js.write("\"document\":{\n")
    exp="^PN (.*)"
    js.write("\t\"PN\":"+re.findall(exp,d[0])[0]+",\n")
    exp="\nRN (.*)?"
    js.write("\t\"RN\":"+re.findall(exp,d[0])[0]+",\n")    
    exp="\nAN (.*)"
    js.write("\t\"AN\":"+re.findall(exp,d[0])[0]+",\n")
    exp="\nAU (.*)"
    if (len(re.findall(exp,d[0]))>0):
        js.write("\t\"AU\": \""+re.findall(exp,d[0])[0]+"\",\n")
    exp="\nTI ((.|\n|\r\n)*?)\n\w"
    js.write("\t\"TI\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
    exp="\nSO (.*)"
    js.write("\t\"SO\": \""+re.findall(exp,d[0])[0]+"\",\n")
    exp="\nMJ ((.|\n|\r\n)*?)\n\w"
    js.write("\t\"MJ\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
    exp="\nMN ((.|\n|\r\n)*?)\n\w"
    js.write("\t\"MN\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
    
    exp="\nAB ((.|\n|\r\n)*?)\n\w"
    if (len(re.findall(exp,d[0]))>0):
        js.write("\t\"AB\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
        
    exp="\nEX ((.|\n|\r\n)*?)\n\w"
    if (len(re.findall(exp,d[0]))>0):
        js.write("\t\"EX\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
        
    exp="\nRF ((.|\n|\r\n)*?)\n\w"
    if (len(re.findall(exp,d[0]))>0):
        js.write("\t\"RF\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
        print("\t\"RF\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
        
    exp="\nCT ((.|\n|\r\n)*?)\n\w"
    if (len(re.findall(exp,d[0]))>0):
        js.write("\t\"CT\": \""+re.findall(exp,d[0])[0][0]+"\",\n")
        print("\t\"CT\": \""+re.findall(exp,d[0])[0][0]+"\",\n")   
    
    js.write("}\n\n")
    
js.write("}\n")
js.close()



```
