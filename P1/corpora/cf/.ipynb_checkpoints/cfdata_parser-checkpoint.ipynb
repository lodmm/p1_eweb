```python
import re, json

files=["cf74","cf75","cf76","cf77","cf78","cf79"]

npapers=0
recnump=0

for file in files:
    
    f = open(file,"r")
    data = f.read()
    f.close()

    papers = []

    exp="(PN (.|\n|\r\n)*?)\n\n"
    for d in re.findall(exp,data):
        paper={}
        npapers=npapers+1
    
        exp="^PN (.*)"
        paper['paperNum']=int(re.findall(exp,d[0])[0])
        print(paper['paperNum'])
    
        exp="\nRN (\d*)\s+"
        paper['recordNum']=int(re.findall(exp,d[0])[0])
        
        if(paper['recordNum']!=recnump+1):
            print("FALTA 1!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!")
        recnump=paper['recordNum']
    
        exp="\nAN (.*)"
        paper['acessionNum']=int(re.findall(exp,d[0])[0])
    
        exp="\nAU (.*)\."
        if (len(re.findall(exp,d[0]))>0):
            paper['authors']=re.findall(exp,d[0])[0].replace("\n ","").split(".  ")
        
        exp="\nTI ((.|\n|\r\n)*?).\n\w"
        paper['title']=re.findall(exp,d[0])[0][0].replace("\n  ","")
    
        exp="\nSO (.*)"
        paper['source']=re.findall(exp,d[0])[0]
    
        exp="\nMJ ((.|\n|\r\n)*?).\n\w"
        if (len(re.findall(exp,d[0]))>0):
            paper['majorSubjects']=re.findall(exp,d[0])[0][0].replace("\n ","").split(".  ")
    
        exp="\nMN ((.|\n|\r\n)*?).\n\w"
        paper['minorSubjects']=re.findall(exp,d[0])[0][0].replace("\n ","").split(".  ")
    
        exp="\nAB ((.|\n|\r\n)*?)\n\w"
        if (len(re.findall(exp,d[0]))>0):
            result=re.sub("\s*\n\s*"," ",re.findall(exp,d[0])
            paper['abstract/extract']=re.sub(re.findall(exp,d[0])[0][0].replace("\n ","")
        
        exp="\nEX ((.|\n|\r\n)*?)\n\w"
        if (len(re.findall(exp,d[0]))>0):
            paper['abstract_extract']=re.findall(exp,d[0])[0][0].replace("\n ","")
        
        exp="\nRF ((.|\n|\r\n)*?)(\n\w|$)"
        result=re.findall(exp,d[0])
        if (len(result)>0):
            paper['references']=result[0][0].split("\n   ")

        exp="\nCT   ((.|\n|\r\n)*?)$"
        result=re.findall(exp,d[0])
        if (len(result)>0):        
            paper['citations']=re.split("\n\s+",result[0][0])
    
        papers.append(paper)
    

    jsf = open(file+".json","w") 
    jsf.write(json.dumps(papers,sort_keys=False, indent=2, separators=(',', ': ')))
    jsf.close()

print("NUMERO DE PAPERS: ",npapers)
```
